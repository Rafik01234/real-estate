# 1.Посчитаем цену за квадратный метр

```python
df2
```

Общая формула для нахождения цены за кв метр должен быть таким: цена_за_квм = цена_кв / общая_квадратура

Получается чтобы найти цену за один квадратный метр, мы должны просто поделить Цену квартиры на его квадратуру, к примеру кв стоит 5млн рублей, а его квадратура пусть будет 50квм. Получается цена_за_1квм= 5000000 / 50 = 500000 рублей. Но это только для одной квартиры, но ведь в городах квартир много и считать каждую по отдельности занимает много времени, поэтому мы можем просто сгруппировать столбец по городу и дальше уже вывести среднее значение для каждого города. 

```python
df2['price_for_1m'] = df2['last_price'] / df2['total_area']
```

```python
df2.groupby('locality_name').agg({'price_for_1m':'mean'}).sort_values(by = 'price_for_1m', ascending = False).reset_index()
```

Теперь мы можем видеть цену за 1квадратный метр по каждому городу. Самая высокая цена за кв метр это у нас в Санкт-Петербурге, потому что это огромный город и там квартиры дорогие, а самая низкая цена у нас в городе Сланцы. Но эти данные могут быть некорректными, так как мы не убрали выбросы. Так же мы можем посмотреть медиану цены за квадратный метр

```python
df2.groupby('locality_name').agg({'price_for_1m':'median'}).sort_values(by = 'price_for_1m', ascending = False).reset_index()
```

```python
import matplotlib.pyplot as plt
data = sp['days_exposition']
plt.hist(data, bins = 10, color = 'blue', edgecolor = 'black')
plt.xlabel('Количество дней на продаже')
plt.ylabel('Количество объявлений')
plt.show()
```

Мы имеем 15721 объявлений и из них примерно 13000 объявлений находились на сайте около 180 дней, что равняется почти полугодам. Получается объявлений на сайте находятся в среднем полгода.

```python
sp1 = sp.query('0 < rooms < 10')
counts, bins, _ = plt.hist(sp1['rooms'], bins=15, color='green', edgecolor='black')

plt.ylabel('Количество Объявлений')
plt.xlabel('Количество комнат')


for i in range(len(counts)):
    plt.text(bins[i] + (bins[1] - bins[0]) / 2, counts[i], int(counts[i]), 
             ha='center', va='bottom', color='black')

plt.show()
```

```python
sp2 = sp.query('ceiling_height < 4')
counts, bins, _ = plt.hist(sp2['ceiling_height'], bins=15, color='purple', edgecolor='black')

plt.ylabel('Количество Объявлений')
plt.xlabel('Высота потолков')


for i in range(len(counts)):
    plt.text(bins[i] + (bins[1] - bins[0]) / 2, counts[i], int(counts[i]), 
             ha='center', va='bottom', color='black')

plt.show()
```

Мы будем рассматривать квартиры с потолком до 4 метров, так как остальные объявления вряд ли являются квартирами. Средняя высота потолков в городе Санкт-Петербург равняется 2,6-2,7 метрам.
